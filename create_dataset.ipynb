# Recomendation System
In this notebook we create the dataset from web-scraping of imdb website

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Threr are 5000 movies list and the link is given below.
This is the first page of 100 movies list. <br/>
The bs4 library is used to screp the data from website

```python
url = 'https://www.imdb.com/list/ls063676189/'

response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')
```

### Screp the desired tag

```python
name = soup.find_all('img', class_='loadlate')
year = soup.find_all('span', class_='lister-item-year text-muted unbold')
genre = soup.find_all('span', class_ = 'genre')
duration = soup.find_all('span', class_='runtime')
rate = soup.find_all('div', class_='ipl-rating-star small')
# n0 = name[0]
# l = n0.text.split()
```

### Store the data into list

```python
name_list = []
name_id = []
img_url = []
movie_genre = []
released_year = []
m_duration = []
m_rating = []
for n in name:
    try:
        name_id.append(n.attrs['data-tconst'])
        name_list.append(n.attrs['alt'])
        img_url.append(n.attrs['loadlate'])
    except:
        pass
    
for y in year:
    released_year.append(y.text)
for d in duration:
    m_duration.append(d.text)
for span in rate:
    rating = span.find_all('span', class_='ipl-rating-star__rating')
    m_rating.append(rating[0].text)
    
for g in genre:
    movie_genre.append(g.text)
```

```python
len(name_list), len(name_id), len(img_url), len(movie_genre), len(released_year), len(m_duration), len(m_rating)
```

screp the all 49 page 

```python
i = 2
while i<=50:
    urls= 'https://www.imdb.com/list/ls063676189/?sort=list_order,asc&st_dt=&mode=detail&page='+str(i)
    i = i+1
    
    response = requests.get(urls)
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')
    name = soup.find_all('img', class_='loadlate')
    year = soup.find_all('span', class_='lister-item-year text-muted unbold')
    genre = soup.find_all('span', class_ = 'genre')
    duration = soup.find_all('span', class_='runtime')
    rate = soup.find_all('div', class_='ipl-rating-star small')
    
    for n in name:
        try:
            name_id.append(n.attrs['data-tconst'])
            name_list.append(n.attrs['alt'])
            img_url.append(n.attrs['loadlate'])
        except:
            pass
    for y in year:
        released_year.append(y.text)
        
    for d in duration:
        m_duration.append(d.text)
        
    for span in rate:
        rating = span.find_all('span', class_='ipl-rating-star__rating')
        m_rating.append(rating[0].text)
    
    for g in genre:
        movie_genre.append(g.text)
```

### Create the dataframe of these data

```python
import pandas as pd
imdb_movies_data = pd.DataFrame({
                        'movies_id':name_id,
                        'movies_name':name_list,
                        'genre':movie_genre,
                        'released_year':released_year,
                        'duration':m_duration,
                        'rating':m_rating,
                        'Image_url':img_url
                        })
```

```python
imdb_movies_data
```

### prepare the genre column and capitalize the movies name

```python
imdb_movies_data['genres'] = imdb_movies_data['genre'].apply(lambda x: x.replace('\n', '').replace('            ', '').replace(', ', ' '))
```

```python
imdb_movies_data['movies_name'] = imdb_movies_data['movies_name'].str.capitalize()
```

```python
imdb_movies_data.drop(['genre'], axis = 1, inplace = True)
```
